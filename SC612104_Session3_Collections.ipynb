## 1. ทำไมต้องมี Collection?

จนถึงตอนนี้ ตัวแปรที่เราใช้เก็บได้แค่ **1 ค่าต่อตัวแปร** เช่น `name = "มานี"` เก็บชื่อได้คนเดียว

ลองคิดปัญหานี้: ถ้าต้องเก็บชื่อนักศึกษาทั้งห้อง 40 คน จะสร้างตัวแปร 40 ตัวไหม?

```python
# วิธีแย่มาก - ไม่ทำตามนี้!
student1 = "มานี"
student2 = "สมชาย"
student3 = "วิชัย"
# ... ต้องเขียนต่ออีก 37 บรรทัด
```

**Collection** คือชนิดข้อมูลที่เก็บ "หลายค่า" ไว้ในตัวแปรเดียว Python มี 4 collection หลักที่ต้องรู้จัก:

| Collection | เก็บข้อมูลแบบ | เปลี่ยนแปลงได้ไหม | มีลำดับไหม | ซ้ำกันได้ไหม |
|---|---|---|---|---|
| `list` | ลำดับของค่า | ✅ แก้ไขได้ (mutable) | ✅ มีลำดับ | ✅ ซ้ำได้ |
| `tuple` | ลำดับของค่า | ❌ แก้ไขไม่ได้ (immutable) | ✅ มีลำดับ | ✅ ซ้ำได้ |
| `set` | กลุ่มของค่าที่ไม่ซ้ำ | ✅ แก้ไขได้ | ❌ ไม่มีลำดับ | ❌ ไม่ซ้ำ |
| `dict` | คู่ key-value | ✅ แก้ไขได้ | ✅ มีลำดับ (Python 3.7+) | key ไม่ซ้ำ, value ซ้ำได้ |

จะเห็นความแตกต่างชัดเจนขึ้นเรื่อย ๆ ในแต่ละหัวข้อด้านล่าง

## 2. list

**list** คือ collection ที่เก็บค่าหลายค่าเรียงเป็นลำดับ แก้ไขได้ และมีค่าซ้ำกันได้ ใช้เครื่องหมาย `[ ]` ในการสร้าง

### 2.1 การสร้าง list


````python
# สร้าง list ว่าง
empty_list = []

# สร้าง list ที่มีข้อมูล
fruits = ["แอปเปิ้ล", "กล้วย", "ส้ม", "มะม่วง"]
numbers = [10, 20, 30, 40, 50]

# list เก็บข้อมูลคนละชนิดกันได้ในตัวเดียว (ไม่นิยมทำ แต่ทำได้)
mixed = ["มานี", 20, 3.75, True]

print(fruits)
print(numbers)
print(mixed)
print(type(fruits))
````

### 2.2 การเข้าถึงสมาชิกใน list (Indexing)

ทุกสมาชิกใน list มี "ตำแหน่ง" หรือ index กำกับอยู่ — **Python เริ่มนับ index จาก 0 เสมอ** ไม่ใช่ 1!

```
index:    0       1       2       3
list:  ["แอปเปิ้ล", "กล้วย", "ส้ม", "มะม่วง"]
neg-idx: -4      -3      -2      -1
```

นอกจากนับจากซ้าย (0, 1, 2, ...) ยังนับจากขวาด้วย index ลบได้ (-1 คือตัวสุดท้าย)


````python
fruits = ["แอปเปิ้ล", "กล้วย", "ส้ม", "มะม่วง"]

print(fruits[0])    # แอปเปิ้ล (ตัวแรก)
print(fruits[2])    # ส้ม
print(fruits[-1])   # มะม่วง (ตัวสุดท้าย)
print(fruits[-2])   # ส้ม (ตัวรองสุดท้าย)

# ถ้า index เกินขอบเขต จะเกิด error
# print(fruits[10])   # ลบ # แล้วรันดู จะเห็น IndexError
````

### 2.3 Slicing — ตัดเอาช่วงของ list

ใช้ `list[start:stop]` เพื่อตัดเอาช่วงของสมาชิก (ไม่รวม `stop` เหมือนกับ `range()`)


````python
numbers = [10, 20, 30, 40, 50, 60, 70]

print(numbers[1:4])     # [20, 30, 40] -- index 1,2,3 (ไม่รวม 4)
print(numbers[:3])      # [10, 20, 30] -- จากต้นจนถึง index 2
print(numbers[3:])      # [40, 50, 60, 70] -- จาก index 3 ถึงท้าย
print(numbers[:])       # ทั้งหมด (มักใช้ทำ copy ของ list)
print(numbers[::2])     # [10, 30, 50, 70] -- ก้าวกระโดดทีละ 2
print(numbers[::-1])    # กลับลำดับ list ทั้งหมด
````

### 2.4 การแก้ไข list (Mutability)

list แก้ไขค่าได้โดยตรง ผ่าน index หรือเมธอดต่างๆ


````python
fruits = ["แอปเปิ้ล", "กล้วย", "ส้ม"]

# แก้ไขค่าโดยตรงผ่าน index
fruits[1] = "สับปะรด"
print(fruits)   # ['แอปเปิ้ล', 'สับปะรด', 'ส้ม']
````

### 2.5 เมธอดสำคัญของ list

| เมธอด | ความหมาย | ตัวอย่าง |
|---|---|---|
| `.append(x)` | เพิ่ม x ไปท้าย list | `fruits.append("ฝรั่ง")` |
| `.insert(i, x)` | แทรก x ที่ตำแหน่ง i | `fruits.insert(1, "ฝรั่ง")` |
| `.remove(x)` | ลบค่า x ตัวแรกที่เจอ | `fruits.remove("กล้วย")` |
| `.pop(i)` | ลบและคืนค่าที่ตำแหน่ง i (ไม่ระบุ = ตัวสุดท้าย) | `fruits.pop()` |
| `.sort()` | เรียงลำดับ list (แก้ไขตัวเดิม) | `numbers.sort()` |
| `.reverse()` | กลับลำดับ list (แก้ไขตัวเดิม) | `numbers.reverse()` |
| `len(list)` | จำนวนสมาชิกใน list | `len(fruits)` |
| `x in list` | เช็คว่า x อยู่ใน list ไหม | `"กล้วย" in fruits` |


````python
fruits = ["แอปเปิ้ล", "กล้วย", "ส้ม"]

fruits.append("มะม่วง")
print("หลัง append:", fruits)

fruits.insert(1, "ฝรั่ง")
print("หลัง insert:", fruits)

fruits.remove("กล้วย")
print("หลัง remove:", fruits)

last_fruit = fruits.pop()
print("ตัวที่ pop ออกมา:", last_fruit)
print("หลัง pop:", fruits)

print("จำนวนสมาชิก:", len(fruits))
print("มีส้มไหม:", "ส้ม" in fruits)
````

````python
numbers = [5, 2, 8, 1, 9, 3]

numbers.sort()
print("เรียงน้อยไปมาก:", numbers)

numbers.sort(reverse=True)
print("เรียงมากไปน้อย:", numbers)

numbers.reverse()
print("กลับลำดับ (จากที่เรียงมากไปน้อย):", numbers)
````

### 2.6 list ซ้อน list (Nested List) — เก็บข้อมูล 2 มิติ

list เก็บ list ไว้ข้างในได้ — ใช้บ่อยมากเวลาเก็บข้อมูลแบบตาราง (แถว x คอลัมน์)


````python
# ตารางคะแนนนักศึกษา 3 คน วิชาละ 3 คะแนน (แถว x คอลัมน์)
scores = [
    [80, 90, 75],   # นักศึกษาคนที่ 1
    [60, 70, 65],   # นักศึกษาคนที่ 2
    [95, 88, 92],   # นักศึกษาคนที่ 3
]

print(scores[0])       # คะแนนทั้งหมดของนักศึกษาคนที่ 1
print(scores[0][1])    # คะแนนวิชาที่ 2 ของนักศึกษาคนที่ 1
print(scores[2][0])    # คะแนนวิชาที่ 1 ของนักศึกษาคนที่ 3
````

## 3. tuple

**tuple** คือ collection ที่เก็บค่าหลายค่าเรียงเป็นลำดับ **เหมือน list ทุกอย่าง ยกเว้นแก้ไขไม่ได้** (immutable) ใช้เครื่องหมาย `( )` ในการสร้าง

### 3.1 การสร้าง tuple


````python
# สร้าง tuple
coordinates = (13.736717, 100.523186)   # ละติจูด, ลองจิจูด ของกรุงเทพฯ
rgb_color = (255, 0, 0)                  # สีแดงในรูปแบบ RGB

print(coordinates)
print(rgb_color)
print(type(coordinates))

# tuple ที่มีสมาชิกตัวเดียว ต้องมี , ต่อท้าย ไม่งั้นจะไม่ใช่ tuple!
not_a_tuple = (5)        # นี่คือ int ธรรมดา ไม่ใช่ tuple
single_tuple = (5,)      # ต้องมี comma ต่อท้าย จึงเป็น tuple
print(type(not_a_tuple))   # <class 'int'>
print(type(single_tuple))  # <class 'tuple'>
````

### 3.2 การเข้าถึงสมาชิกใน tuple

ใช้ index และ slicing แบบเดียวกับ list ได้ทุกอย่าง — ต่างกันแค่**แก้ไขค่าไม่ได้**


````python
coordinates = (13.736717, 100.523186)

print(coordinates[0])    # 13.736717
print(coordinates[1])    # 100.523186

# ลองแก้ไขค่า -> จะเกิด error เพราะ tuple แก้ไขไม่ได้
# coordinates[0] = 14.0   # ลบ # แล้วรันดู จะเห็น TypeError
````

### 3.3 Tuple Unpacking — แตกค่าจาก tuple ใส่ตัวแปรหลายตัว

เป็นการใช้งานที่นิยมมากของ tuple — เอาแต่ละค่าของ tuple ไปใส่ตัวแปรในบรรทัดเดียว


````python
coordinates = (13.736717, 100.523186)

latitude, longitude = coordinates   # unpack ทั้ง 2 ค่าในบรรทัดเดียว
print(f"ละติจูด: {latitude}, ลองจิจูด: {longitude}")

# ใช้บ่อยมากตอน return ค่าหลายค่าจากฟังก์ชัน (function จะเรียนละเอียดในคาบถัดไป)
def get_min_max(numbers):
    return min(numbers), max(numbers)   # จริงๆ คือ return tuple (min, max)

result = get_min_max([5, 2, 8, 1, 9])
print(result)          # (1, 9) -- ได้ tuple กลับมา
low, high = result      # unpack ได้เลย
print(f"ต่ำสุด: {low}, สูงสุด: {high}")
````

### 3.4 เมื่อไหร่ใช้ tuple แทน list?

| ใช้ `list` เมื่อ... | ใช้ `tuple` เมื่อ... |
|---|---|
| ข้อมูลต้องเปลี่ยนแปลงได้ (เพิ่ม/ลบ/แก้) | ข้อมูลไม่ควรถูกแก้ไขเลย (เช่น พิกัด, ค่าคงที่) |
| ลำดับรายการที่ขยาย/หดได้ | กลุ่มค่าที่มีจำนวนคงที่และความหมายชัดเจนตามตำแหน่ง |
| เช่น รายชื่อนักศึกษาที่เพิ่ม/ลบได้ | เช่น (วัน, เดือน, ปี) หรือ (R, G, B) |

**ข้อดีเสริม:** tuple ทำงานเร็วกว่า list เล็กน้อย และใช้เป็น key ของ dict ได้ (list ใช้เป็น key ไม่ได้ เพราะแก้ไขได้)


````python
# ตัวอย่าง: ใช้ tuple เป็น key ของ dict (list ทำแบบนี้ไม่ได้!)
distances = {
    (13.7, 100.5): "กรุงเทพฯ",
    (18.8, 98.9): "เชียงใหม่",
}
print(distances[(13.7, 100.5)])
````

## 4. set

**set** คือ collection ที่เก็บค่าที่ **ไม่ซ้ำกัน** และ **ไม่มีลำดับ** ใช้เครื่องหมาย `{ }` ในการสร้าง (ต้องมีสมาชิกข้างใน ไม่งั้นจะกลายเป็น dict ว่าง)

### 4.1 การสร้าง set


````python
# สร้าง set
fruits_set = {"แอปเปิ้ล", "กล้วย", "ส้ม", "แอปเปิ้ล"}   # ใส่ "แอปเปิ้ล" ซ้ำ 2 ครั้ง
print(fruits_set)   # จะเหลือแค่ตัวเดียว เพราะ set ไม่เก็บค่าซ้ำ
print(type(fruits_set))

# สร้าง set ว่าง - ต้องใช้ set() ไม่ใช่ {} (เพราะ {} คือ dict ว่าง)
empty_set = set()
print(type(empty_set))

# แปลง list เป็น set เพื่อลบค่าซ้ำ (ใช้บ่อยมาก!)
numbers_with_duplicates = [1, 2, 2, 3, 3, 3, 4]
unique_numbers = set(numbers_with_duplicates)
print(unique_numbers)   # {1, 2, 3, 4}
````

### 4.2 การเพิ่ม/ลบสมาชิกใน set

**ข้อสำคัญ:** set ไม่มีลำดับ จึงเข้าถึงสมาชิกผ่าน index แบบ `fruits_set[0]` ไม่ได้ (จะ error) ต้องใช้ loop หรือเช็คด้วย `in` แทน


````python
fruits_set = {"แอปเปิ้ล", "กล้วย", "ส้ม"}

fruits_set.add("มะม่วง")          # เพิ่มสมาชิก
print("หลัง add:", fruits_set)

fruits_set.remove("กล้วย")        # ลบสมาชิก (error ถ้าไม่มีอยู่จริง)
print("หลัง remove:", fruits_set)

fruits_set.discard("ทุเรียน")     # ลบสมาชิก (ไม่ error แม้ไม่มีอยู่จริง - ปลอดภัยกว่า remove)
print("หลัง discard (ไม่มีทุเรียนอยู่แล้ว แต่ไม่ error):", fruits_set)

print("มีส้มไหม:", "ส้ม" in fruits_set)
print("จำนวนสมาชิก:", len(fruits_set))
````

### 4.3 Set Operations — การคำนวณทางคณิตศาสตร์ของเซต

จุดเด่นที่สุดของ set คือการคำนวณแบบเซตในคณิตศาสตร์ (Union, Intersection, Difference) ซึ่งมีประโยชน์มากในการวิเคราะห์ข้อมูล เช่น หาความเหมือน/ต่างของ 2 กลุ่ม

| Operation | สัญลักษณ์ | ความหมาย | เมธอด |
|---|---|---|---|
| Union | \| | รวมสมาชิกทั้งสองเซต (ไม่ซ้ำ) | `.union()` |
| Intersection | & | สมาชิกที่มีอยู่ใน**ทั้งสอง**เซต | `.intersection()` |
| Difference | - | สมาชิกที่มีใน A แต่ไม่มีใน B | `.difference()` |
| Symmetric Difference | ^ | สมาชิกที่อยู่ในเซตใดเซตหนึ่งเท่านั้น (ไม่ใช่ทั้งคู่) | `.symmetric_difference()` |


````python
# ตัวอย่าง: นักศึกษาที่ลงทะเบียนวิชา A และวิชา B
students_course_a = {"มานี", "สมชาย", "วิชัย", "ปรีชา"}
students_course_b = {"สมชาย", "วิชัย", "สมหญิง", "อรุณ"}

# Union: นักศึกษาทั้งหมดที่ลงเรียนวิชาใดวิชาหนึ่ง (หรือทั้งสอง)
print("Union:", students_course_a | students_course_b)

# Intersection: นักศึกษาที่ลงทั้ง 2 วิชา
print("Intersection:", students_course_a & students_course_b)

# Difference: นักศึกษาที่ลงวิชา A แต่ไม่ได้ลงวิชา B
print("Difference (A-B):", students_course_a - students_course_b)

# Symmetric Difference: นักศึกษาที่ลงแค่วิชาเดียว (ไม่ใช่ทั้งคู่)
print("Symmetric Difference:", students_course_a ^ students_course_b)
````

### 4.4 เมื่อไหร่ใช้ set?

- ต้องการ**ลบค่าซ้ำ**ออกจากข้อมูล (วิธีที่เร็วและง่ายที่สุด)
- ต้องการเช็คว่า "มีอยู่ไหม" บ่อยๆ (set เช็ค `in` เร็วกว่า list มาก เมื่อข้อมูลมีจำนวนมาก)
- ต้องการเปรียบเทียบกลุ่มข้อมูล 2 กลุ่ม (เหมือน/ต่างกันตรงไหน)
- **ไม่เหมาะ** เมื่อต้องการรักษาลำดับ หรือมีค่าซ้ำที่มีความหมาย

## 5. dict

**dict** (dictionary) คือ collection ที่เก็บข้อมูลเป็นคู่ **key-value** ใช้เครื่องหมาย `{ }` พร้อม `:` คั่นระหว่าง key กับ value

### 5.1 การสร้าง dict


````python
# สร้าง dict ว่าง
empty_dict = {}

# สร้าง dict ที่มีข้อมูล - key: value
student = {
    "name": "มานี",
    "age": 20,
    "gpa": 3.75,
    "is_active": True
}

print(student)
print(type(student))
````

### 5.2 การเข้าถึงค่าใน dict

เข้าถึงค่าผ่าน **key** (ไม่ใช่ index แบบ list/tuple) ใช้ `dict[key]` หรือ `.get(key)` (ปลอดภัยกว่า เพราะไม่ error ถ้าไม่มี key นั้น)


````python
student = {"name": "มานี", "age": 20, "gpa": 3.75}

print(student["name"])    # มานี
print(student["age"])     # 20

# ถ้า key ไม่มีอยู่จริง จะ error
# print(student["email"])    # ลบ # แล้วรันดู จะเห็น KeyError

# ใช้ .get() ปลอดภัยกว่า - ถ้าไม่มี key จะได้ None แทน error
print(student.get("email"))            # None
print(student.get("email", "ไม่พบข้อมูล"))   # กำหนดค่า default เองได้
````

### 5.3 การแก้ไขและเพิ่มข้อมูลใน dict


````python
student = {"name": "มานี", "age": 20}

# แก้ไขค่าเดิม
student["age"] = 21
print("หลังแก้ไข age:", student)

# เพิ่ม key ใหม่ (ถ้า key ไม่มีอยู่ Python จะสร้างให้เอง)
student["gpa"] = 3.75
print("หลังเพิ่ม gpa:", student)

# ลบ key
del student["age"]
print("หลังลบ age:", student)

# .pop() ลบและคืนค่าด้วย
removed_value = student.pop("gpa")
print("ค่าที่ pop ออกมา:", removed_value)
print("หลัง pop gpa:", student)
````

### 5.4 เมธอดสำคัญของ dict

| เมธอด | ความหมาย |
|---|---|
| `.keys()` | คืน key ทั้งหมด |
| `.values()` | คืน value ทั้งหมด |
| `.items()` | คืนคู่ (key, value) ทั้งหมด |
| `key in dict` | เช็คว่ามี key นี้อยู่ไหม |
| `len(dict)` | จำนวนคู่ key-value |

## 6. สรุปเปรียบเทียบ list vs tuple vs set vs dict

| | `list` | `tuple` | `set` | `dict` |
|---|---|---|---|---|
| **สัญลักษณ์สร้าง** | `[1, 2, 3]` | `(1, 2, 3)` | `{1, 2, 3}` | `{"a": 1, "b": 2}` |
| **แก้ไขได้ (mutable)** | ✅ | ❌ | ✅ | ✅ |
| **มีลำดับ (ordered)** | ✅ | ✅ | ❌ | ✅ (key ตามลำดับที่ใส่) |
| **มีค่าซ้ำได้** | ✅ | ✅ | ❌ | key ไม่ซ้ำ, value ซ้ำได้ |
| **เข้าถึงผ่าน** | index `[0]` | index `[0]` | ไม่มี index ตรง ใช้ `in`/loop | key `["ชื่อ"]` |
| **ใช้เมื่อ** | ข้อมูลที่ต้องแก้ไข/ขยาย | ข้อมูลคงที่ ไม่ควรแก้ | ต้องการค่าไม่ซ้ำ/set operations | ข้อมูลที่มีชื่อกำกับ (key-value) |

## 7. Comprehension

**Comprehension** คือวิธีเขียน loop แบบสั้นกระชับ เพื่อสร้าง list/dict/set ใหม่จากของเดิม — เขียนในบรรทัดเดียว แทนการเขียน `for` loop หลายบรรทัด

### 7.1 List Comprehension

**โครงสร้างพื้นฐาน:**
```python
ผลลัพธ์ = [expression for item in iterable]
```

ลองดูปัญหา: อยากได้ list ของเลขยกกำลัง 2 จาก 1 ถึง 5


````python
# วิธีเดิม - ใช้ for loop ปกติ
squares = []
for num in range(1, 6):
    squares.append(num ** 2)
print("วิธีเดิม:", squares)

# วิธีใหม่ - list comprehension (บรรทัดเดียว!)
squares_v2 = [num ** 2 for num in range(1, 6)]
print("list comprehension:", squares_v2)
````

### 7.2 List Comprehension พร้อมเงื่อนไข (filter)

**โครงสร้างที่มีเงื่อนไข:**
```python
ผลลัพธ์ = [expression for item in iterable if เงื่อนไข]
```


````python
# วิธีเดิม - กรองเอาเฉพาะเลขคู่ออกมาจาก 1-20
even_numbers = []
for num in range(1, 21):
    if num % 2 == 0:
        even_numbers.append(num)
print("วิธีเดิม:", even_numbers)

# วิธีใหม่ - list comprehension พร้อมเงื่อนไข
even_numbers_v2 = [num for num in range(1, 21) if num % 2 == 0]
print("list comprehension:", even_numbers_v2)
````

````python
# ตัวอย่างที่ใช้บ่อย: แปลงข้อมูลทั้ง list พร้อมกัน
names = ["มานี", "สมชาย", "วิชัย"]

# แปลงเป็นตัวพิมพ์ใหญ่ทั้งหมด (ใช้กับ .upper() ของ string)
names_upper = [name.upper() for name in names]
print(names_upper)

# หาความยาวของแต่ละชื่อ
name_lengths = [len(name) for name in names]
print(name_lengths)

# กรองเฉพาะชื่อที่ยาวมากกว่า 4 ตัวอักษร
long_names = [name for name in names if len(name) > 4]
print(long_names)
````

### 7.3 Dict Comprehension

**โครงสร้าง:**
```python
ผลลัพธ์ = {key_expr: value_expr for item in iterable}
```


````python
# สร้าง dict จาก list สองตัว (ชื่อ -> คะแนน)
names = ["มานี", "สมชาย", "วิชัย"]
scores = [85, 90, 78]

# วิธีเดิม - for loop
score_dict = {}
for i in range(len(names)):
    score_dict[names[i]] = scores[i]
print("วิธีเดิม:", score_dict)

# วิธีใหม่ - dict comprehension ใช้คู่กับ zip() (เรียนรายละเอียด zip ในหัวข้อถัดไป)
score_dict_v2 = {name: score for name, score in zip(names, scores)}
print("dict comprehension:", score_dict_v2)

# ตัวอย่าง: สร้าง dict เลขยกกำลัง 2 พร้อมกรองเงื่อนไข
squares_dict = {num: num ** 2 for num in range(1, 8) if num % 2 == 0}
print(squares_dict)   # {2: 4, 4: 16, 6: 36}
````

### 7.4 Set Comprehension

**โครงสร้าง:**
```python
ผลลัพธ์ = {expression for item in iterable}
```

หน้าตาคล้าย dict comprehension มาก แต่ไม่มี `:` คั่น key-value (เพราะ set ไม่มี key-value)


````python
# หาความยาวที่ไม่ซ้ำกันของแต่ละชื่อ
names = ["มานี", "สมชาย", "วิชัย", "อร", "ปรีชา"]

unique_lengths = {len(name) for name in names}
print(unique_lengths)   # set ของความยาวที่ไม่ซ้ำกัน

# เทียบกับถ้าใช้ list comprehension (จะมีค่าซ้ำ)
all_lengths = [len(name) for name in names]
print(all_lengths)      # list ที่มีค่าซ้ำได้
````

### 7.5 เมื่อไหร่ควร (และไม่ควร) ใช้ Comprehension

**ควรใช้เมื่อ:** logic ง่าย กระชับ อ่านแล้วเข้าใจได้ในแวบแรก

**ไม่ควรใช้เมื่อ:** logic ซับซ้อนมาก มี condition ซ้อนหลายชั้น — เขียนเป็น `for` loop ปกติจะอ่านง่ายกว่า

```python
# อ่านยาก เกินไป! ไม่ควรเขียนแบบนี้
result = [x**2 if x % 2 == 0 else x**3 for x in range(20) if x > 5 if x < 15]

# เขียนแบบนี้อ่านง่ายกว่ามาก
result = []
for x in range(20):
    if x > 5 and x < 15:
        if x % 2 == 0:
            result.append(x ** 2)
        else:
            result.append(x ** 3)
```

> 💡 **กฎง่าย ๆ:** ถ้า comprehension ของคุณยาวเกิน 1 บรรทัดปกติ หรือต้องอ่าน 2-3 รอบกว่าจะเข้าใจ — เปลี่ยนไปเขียน `for` loop ธรรมดาจะดีกว่า

## 8. การวนลูปผ่าน Collection

### 8.1 วนลูปผ่าน list และ tuple

วนตรงๆ ได้เลย จะได้สมาชิกแต่ละตัวตามลำดับ (วิธีเดียวกันทั้ง list และ tuple)


````python
fruits = ["แอปเปิ้ล", "กล้วย", "ส้ม"]

for fruit in fruits:
    print(fruit)

print()

coordinates = (13.7, 100.5, 250)   # ละติจูด, ลองจิจูด, ความสูงจากระดับน้ำทะเล
for value in coordinates:
    print(value)
````

### 8.2 วนลูปผ่าน list/tuple พร้อม index ด้วย `enumerate()`

ถ้าต้องการทั้ง **index** และ **ค่า** ไปพร้อมกัน ใช้ `enumerate()` — ดีกว่าการใช้ `range(len(list))` แล้วเข้าถึงด้วย index มาก เพราะอ่านง่ายกว่า


````python
fruits = ["แอปเปิ้ล", "กล้วย", "ส้ม"]

# วิธีที่ไม่แนะนำ (อ่านยาก)
for i in range(len(fruits)):
    print(f"{i}: {fruits[i]}")

print()

# วิธีที่แนะนำ - ใช้ enumerate()
for i, fruit in enumerate(fruits):
    print(f"{i}: {fruit}")

print()

# enumerate() กำหนดเลขเริ่มต้นได้ด้วย (ปกติเริ่ม 0)
for i, fruit in enumerate(fruits, start=1):
    print(f"ลำดับที่ {i}: {fruit}")
````

### 8.3 วนลูปผ่าน set

วนได้เหมือน list แต่**ลำดับที่ได้ไม่แน่นอน** (เพราะ set ไม่มีลำดับ) — ห้ามเขียนโค้ดที่พึ่งพาลำดับการวนของ set


````python
fruits_set = {"แอปเปิ้ล", "กล้วย", "ส้ม"}

for fruit in fruits_set:
    print(fruit)

# หมายเหตุ: ถ้ารันเซลล์นี้หลายรอบ ลำดับที่ print อาจไม่เหมือนกันทุกครั้ง
# (ขึ้นกับ Python version — Python รุ่นใหม่มักจะคงที่ในรอบการรันเดียวกัน แต่ไม่ควรเขียนโค้ดที่พึ่งพาลำดับนี้)
````

### 8.4 วนลูปผ่าน dict — 3 รูปแบบที่ต้องรู้

มี 3 วิธีหลักในการวน dict ขึ้นอยู่กับว่าต้องการ key, value, หรือทั้งคู่


````python
student = {"name": "มานี", "age": 20, "gpa": 3.75}

# วิธี 1: วนผ่าน key เท่านั้น (ค่า default ถ้าไม่ระบุอะไร)
print("--- วนผ่าน key ---")
for key in student:
    print(key)

print()

# วิธี 2: วนผ่าน value เท่านั้น
print("--- วนผ่าน value ---")
for value in student.values():
    print(value)

print()

# วิธี 3: วนผ่านทั้ง key และ value พร้อมกัน (ใช้บ่อยที่สุด!)
print("--- วนผ่าน key และ value ---")
for key, value in student.items():
    print(f"{key}: {value}")
````

### 8.5 วนลูปผ่าน nested dict (โครงสร้างซับซ้อนขึ้น)

ตัวอย่างที่ใช้บ่อยมากในการทำงานกับข้อมูลจริง (เช่นข้อมูลจาก API ที่มักเป็น nested dict)


````python
students = {
    "65010001": {"name": "มานี", "gpa": 3.75},
    "65010002": {"name": "สมชาย", "gpa": 3.20},
    "65010003": {"name": "วิชัย", "gpa": 2.95},
}

for student_id, info in students.items():
    print(f"รหัส {student_id}: {info['name']} (GPA {info['gpa']})")
````

### 8.6 วนลูปผ่าน 2 collection พร้อมกันด้วย `zip()`

`zip()` ใช้ "รวม" หลาย list/tuple เข้าด้วยกันเป็นคู่ๆ ตามลำดับ — มีประโยชน์มากเวลามีข้อมูลหลาย list ที่สัมพันธ์กัน


````python
names = ["มานี", "สมชาย", "วิชัย"]
scores = [85, 90, 78]
grades = ["B+", "A", "C+"]

# zip() รวม 3 list เข้าด้วยกัน วนพร้อมกันทีละชุด
for name, score, grade in zip(names, scores, grades):
    print(f"{name}: คะแนน {score} เกรด {grade}")
````

**ข้อควรระวัง:** ถ้า list ที่ `zip()` มีความยาวไม่เท่ากัน — `zip()` จะหยุดตามตัวที่สั้นที่สุด (ไม่ error แต่อาจทำข้อมูลขาดไปโดยไม่รู้ตัว)


````python
short_list = [1, 2, 3]
long_list = ["a", "b", "c", "d", "e"]

for num, letter in zip(short_list, long_list):
    print(num, letter)
# สังเกตว่า "d" และ "e" จะไม่ถูกใช้เลย เพราะ short_list หมดก่อน
````
